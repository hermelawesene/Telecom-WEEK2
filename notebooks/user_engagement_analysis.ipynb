```python
engagement_analysis.normalize_metrics()
engagement_analysis.run_kmeans()
engagement_analysis.plot_clusters()
cluster_analysis = engagement_analysis.analyze_clusters()
print(cluster_analysis)
```

Interpretation:
Cluster 0: Represents customers with moderate engagement—moderate session duration, frequency, and data usage.
Cluster 1: Represents the most engaged customers—longer session duration, higher session frequency, and higher data usage.
Cluster 2: Represents the least engaged customers—shorter session duration, lowest session frequency, and lowest data usage.
This clustering helps identify different levels of customer engagement, which can inform targeted marketing strategies, customer support prioritization, or product development focused on different user segments. For instance, customers in Cluster 1 might be more loyal or high-value customers, while those in Cluster 2 might need incentives to increase their engagement.
